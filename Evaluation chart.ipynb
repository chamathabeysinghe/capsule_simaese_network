```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ways = np.arange(1,40,2)
```

```python
ways
```

```python
convolutiona_accs = [100.0, 96.2, 93.6, 90.2, 88.0, 88.6, 84.4, 85.0, 86.0, 83.0, 82.6, 82.0, 76.6, 76.2, 74.6, 77.2, 71.0, 72.6, 73.0, 71.4]
capsule_rev2_accs = [100.0, 95.0, 94.6, 91.0, 86.2, 85.8, 81.4, 81.4, 83.0, 78.4, 80.4, 77.2, 78.2, 74.8, 77.0, 74.4, 73.4, 72.2, 70.8, 71.6]
nn_accs = [100.0, 59.8, 46.6, 39.6, 32.0, 31.8, 33.8, 27.6, 32.0, 23.4, 25.0, 22.4, 25.4, 22.0, 20.8, 20.8, 20.0, 20.8, 18.8, 19.0]
random_accs = [100/i for i in ways]
```

```python
plt.figure(figsize=(200,200))
ig,ax = plt.subplots(1)
ax.plot(ways, convolutiona_accs, "m", label="Convolutional")
ax.plot(ways, capsule_rev2_accs, "y", label="Capsule")
plt.plot(ways, nn_accs, label="Nearest neighbour")

ax.plot(ways, random_accs, "g", label="Random guessing")
plt.xlabel("Number of possible classes in one-shot tasks")
plt.ylabel("% Accuracy")
plt.title("Omiglot One-Shot Learning Performance of a Siamese Network")
# box = ax.get_position()
# ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('graph.png', bbox_inches='tight')
plt.show()
```
